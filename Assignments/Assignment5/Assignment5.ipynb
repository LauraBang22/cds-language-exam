### Which assignment generated the most emissions in terms of CO₂eq? Explain why this might be?

I first load the dataframe that contains the CO₂eq per script. 

```python
filename = os.path.join("emissions", "examples", "emissions.csv")
data = pd.read_csv(filename)
```

```python
#defining colorpalettes to use in the plots
colors1 = ["#C9E4DE", "#F7D9C4", "#F7D9C4", "#F7D9C4", "#F2C6DE", "#C6DEF1", "#C6DEF1"]
colors2 = ["#C9E4DE", "#F7D9C4", "#F2C6DE", "#C6DEF1"]
```

First I plot the emission for all the different scripts, I have written for the previous four assignments.

```python
plt.figure(figsize=(8, 6))  
plt.bar(data["project_name"], data["emissions"], color=colors1)

plt.xlabel("scripts")
plt.ylabel("emissions")
plt.title('Distribution of values in emissions across all scripts')

plt.xticks(rotation=65)  
plt.tight_layout()
plt.show()
```

The plot shows that the script that has the highest emissions, is the the script called **assignment4_data**. That makes sense, because that is the script that is processing the Game of Thrones data. That is the biggest dataset we are working with in these assignments, with over 23.000 line in the .csv file, that the classifier has to run on. It is also the script that takes the longest to run. Therefore it explains why that is the script with the higest CO₂eq.

While it is very easy to see just from this plot, that it is assignment 4 that has the highest CO₂eq, I have still combined the emissions from the assignments that have multiple scripts, in the following part.

```python
#combined emissions for assignment 2 and assignment 4
assignment2 = data["emissions"][1]+data["emissions"][2]+data["emissions"][3] 
assignment4 = data["emissions"][5]+data["emissions"][6] 

#created new lists to create a plot from
combined_data = [data["emissions"][0], assignment2, data["emissions"][4], assignment4] 
assignment_name = ["assignment1", "assignment2", "assignment3", "assignment4"]
```

```python
plt.figure(figsize=(8, 6))  
plt.bar(assignment_name, combined_data, color=colors2)

plt.xlabel("assignment")
plt.ylabel("emissions")
plt.title('Distribution of values in emissions across all assignments')


plt.xticks(rotation=0)  
plt.tight_layout()
plt.show()
```

As predicted, this plot shows that the assignment with the highest emission is assignment 4. After that is assignment 1, which still has a higher emission than assignment 2, even though assignment 2 is three different scripts combined. The assignment with the lowest emission is assignment 3, but that is also just running the code once. If you were interested in learning about which words multiple artists use, I think it would quickly add up.

To compare all the scripts from the assignment and to fully be able to tell which task across all three, that have the higest emission I have decided to plot all of them together.

```python
colors3 = ['#97ECF1'] * 8 + ['#FEC868'] * 8 + ['#F7F7F7'] * 2
```

```python
data_assignment2 = pd.concat([data_logistic, data_neural, data_vectorizer])
```

```python
plt.figure(figsize=(12, 6)) 
data_assignment2.plot(x="task_name", y="emissions", kind = "bar", color = colors3, legend=None)

plt.xlabel("tasks")
plt.ylabel("emissions")
plt.title('Distribution of values in emissions across all tasks in assignment 2')

plt.xticks(rotation=85) 
plt.tight_layout()
plt.show()
```

The different colours in the plot represents the different scripts. As can be seen in this plot it is the "fitting" tasks, that generate the most emission. Those are the tasks in the script, where the vectorizer is being fitted to the data and a classifier is being trained. It is the fitting and training of the neural network classifier, that generates the very most CO₂eq. Creating the cross validation plots in the script with the logistic regression classifier also emits high amounts of CO₂eq. Creating the vectorizer by it self in the third script in the task, does not generate very much CO₂eq, which is why a third colour is not visible in the plot. It is these task the emit the most CO₂eq, because they require the computer to handle a lot of data and do a lot of calculations.
