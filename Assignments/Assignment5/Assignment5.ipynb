```python
import os
import sys
sys.path.append("..")
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename = os.path.join("emissions", "emissions.csv")
data = pd.read_csv(filename)
```

```python
data = data.drop(index=2)
```

```python
data
```

```python
colors =["#F8B2BC", "#FBB39A", "#FBB39A", "#FBB39A", "#A6C7EA", "#DFCDE3", "#DFCDE3"]
```

```python
plt.figure(figsize=(8, 6))  # Adjust figure size as needed
plt.bar(data["project_name"], data["emissions"], color=colors)

# Add labels and title
plt.xlabel("tasks")
plt.ylabel("emissions")
plt.title('Distribution of Values in {}'.format("emissions"))


# Display the plot
plt.xticks(rotation=75)  # Rotate x-axis labels for better readability if many categories
plt.tight_layout()
plt.show()
```
